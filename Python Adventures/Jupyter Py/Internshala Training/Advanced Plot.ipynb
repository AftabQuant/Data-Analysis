```python
import matplotlib.pyplot as plt

# Example data
categories = ['A', 'B', 'C', 'D', 'E']  # Categories for the x-axis
values = [23, 45, 56, 78, 44]          # Corresponding values for the y-axis

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue', edgecolor='black')

# Add labels, title, and grid
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the chart
plt.savefig("Bar Chart.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x, y = np.meshgrid(np.arange(-2, 3, 1), np.arange(-2, 3, 1))
u = -y
v = x
plt.quiver(x, y, u, v, angles='xy', scale_units='xy', scale=1, color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quiver Plot Example')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid()
plt.savefig("Quiver Plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='viridis') 
plt.colorbar(label='Function Value') 
plt.title("Meshgrid Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.savefig("MeshGrid Chart.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
x = np.linspace(-5, 5, 100) 
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=10, cmap='plasma')
plt.clabel(contour, inline=True, fontsize=8, fmt="%.2f") 
plt.title("Contour Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(contour, label='Function Value') 
plt.grid(True)
plt.savefig("Contour Plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(8, 6))
plt.fill_between(x, y1, color='blue', alpha=0.5, label='sin(x)')  # Area for sin(x)
plt.fill_between(x, y2, color='orange', alpha=0.5, label='cos(x)')  # Area for cos(x)
plt.title("Area Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc="upper right") 
plt.grid(True)
plt.savefig("Area.png", dpi=300, bbox_inches='tight')
plt.show()
```
