```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("D:\\Ultimate Programming\\Data Bases\\Student Scores.csv")
df = pd.DataFrame(data)
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["EthnicGroup"] = df["EthnicGroup"].replace(np.nan, "Group x")
df["EthnicGroup"].isnull().sum()
```

```python
gp_pe = df.groupby("ParentEduc")["Sl No."].count()
df.fillna({"ParentEduc" :"Unknown"}, inplace=True)
```

```python
gp_tp = df.groupby("TestPrep")["Sl No."].count()
df.fillna({"TestPrep":"Uncompleted"}, inplace=True)
df["TestPrep"].isnull().sum()
```

```python
df["ParentMaritalStatus"].bfill(inplace = True)
df["ParentMaritalStatus"].isnull().sum()
```

```python
df.isnull().sum()
```

```python
gp_ps = df.groupby("PracticeSport")["Sl No."].count()
gp_ps
```

```python
df.fillna({"PracticeSport" : "Unknown"}, inplace=True)
```

```python
df["IsFirstChild"].ffill(inplace = True)
```

```python
df["NrSiblings"] = df["NrSiblings"].replace(np.nan, np.mean(df["NrSiblings"]))
print(np.mean(df["NrSiblings"]))
df["NrSiblings"].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.fillna({"TransportMeans" : "Personal"}, inplace = True)
df["TransportMeans"].isnull().sum()
```

```python
gp_wsh = df.groupby("WklyStudyHours")["Sl No."].count()
gp_wsh
```

```python
df['WklyStudyHours'] = pd.to_numeric(df['WklyStudyHours'].replace({
    '05-Oct': 1,
    '< 5': 3,
    '> 10': 12
}), errors='coerce')
```

```python
gp_wsh = df.groupby("WklyStudyHours")["Sl No."].count()
gp_wsh
```

```python
print(df["WklyStudyHours"].mean())
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].replace(np.nan, 3.4)
gp_wsh = df.groupby("WklyStudyHours")["Sl No."].count()
gp_wsh
```

```python
df.isnull().sum()
```

                Data Visualization Using Different Plots

```python
sns.countplot(data = df, x="EthnicGroup", hue = "Gender", palette = "viridis", order=["group A", "group B", "group C","group D", "group E", "Group x"])
plt.show()
```

```python
sns.barplot(data = df, x = "EthnicGroup", y = "MathScore", estimator="median", hue = "Gender", palette = "coolwarm")
plt.show()
```

```python
sns.lineplot(data = df, x = "EthnicGroup", y = "MathScore", hue = "TestPrep", markers="o")          
plt.show()
```

```python
sns.scatterplot(data = df, x = "ReadingScore", y = "WritingScore")
plt.show()
```

```python
sns.violinplot(data = df, x = "WklyStudyHours")
plt.show()
```

```python
sns.histplot(data = df, x = "NrSiblings", bins = 9, hue = "Gender", palette = "viridis")
plt.show()
```
