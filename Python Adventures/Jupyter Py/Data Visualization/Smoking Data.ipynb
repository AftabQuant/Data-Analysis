```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_excel("Smoking Data.xlsx")
df = pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
sns.barplot(data = df, x = "Year", y = "Smoking_Prevalence", hue = "Gender", estimator='mean', palette = "viridis")
plt.show()
```

```python
sns.countplot(data = df, x = "Year", hue = "Socioeconomic_Status", palette = "viridis")
plt.show()
```

```python
sns.lineplot(data=df, x = "Family_Background", y = "Parental_Supervision", hue = "Gender",palette="viridis")
plt.show()
```

```python
sns.scatterplot(data = df, y = "Smoking_Prevalence", x = "Drug_Experimentation",hue="Gender")
plt.show()
```

```python
sns.histplot(data = df, x = "Mental_Health", hue = "Substance_Education", palette="viridis", kde=True )
plt.show()
```

```python
sns.violinplot(data = df, x = "Smoking_Prevalence", hue="Gender", palette = "viridis")
plt.show()
```

```python
sns.boxplot(data = df, x = "Community_Support", hue = "Socioeconomic_Status", )
plt.show()
```

```python
gp_fb = df.groupby("Family_Background").agg({"Gender" : "count"})
print(gp_fb)
```

```python
sns.heatmap(gp_fb)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.countplot(data=df, x="School_Programs", hue = "Year")
plt.show()
```

```python
sns.countplot(data = df, x = "Age_Group", hue = "Access_to_Counseling", palette="viridis")
plt.show()
```

```python
gp_yr = df.groupby("Year").agg({"Age_Group" : "count"})
print(gp_yr)
```

```python
ex = [0,0,0,0,0.1]
plt.pie(gp_yr["Age_Group"], labels = gp_yr.index, autopct = "%.2f", explode = ex, shadow = True,)
plt.show()
```
