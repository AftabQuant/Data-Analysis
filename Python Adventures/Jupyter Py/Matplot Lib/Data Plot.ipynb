```python
import matplotlib.pyplot as plt
x = ["A", "B", "C", "D", "E", "F"]
y = [20, 15, 10, 15, 20, 16]
plt.bar(x, y, edgecolor = "black")
plt.xlabel("Name")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```

```python
x = ["A", "B", "C", "D", "E", "F"]
y = [20, 15, 10, 15, 20, 16]
z = [32, 20, 18, 10, 12, 2]
plt.plot(x, y, marker = ".", label = "Week1")
plt.plot(x, z, marker = ".", color = "green", label = "Week2")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.figure(figsize=(8,3))
plt.subplot(1,3,2)
plt.plot(x, y2, 'r--')
plt.title('Cosine Wave')

plt.subplot(1,3,1)
plt.plot(x, y1, 'b-')
plt.title('Sine Wave')

plt.subplot(1,3,3)
plt.plot(x, y3, 'b-')
plt.title('Tan Wave')

plt.tight_layout()
plt.show()
```
